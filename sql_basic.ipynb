```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///GoBike.sqlite')
conn = engine.connect()
print(engine.table_names())
```

```python
from sqlalchemy import MetaData, Table
metadata = MetaData()
station = Table('station', metadata, autoload=True, autoload_with=engine)
print (repr(station))
```

```python
from sqlalchemy.orm import Session
import pandas as pd

session = Session(bind=engine)

result = session.query(station)
pd.read_sql(result.statement, engine.connect())
```

```python
rider = Table('rider', metadata, autoload=True, autoload_with=engine)
result2 = session.query(rider)
pd.read_sql(result2.statement, conn)
```

```python
query = 'SELECT * FROM station JOIN rider ON station.station_id = rider.start_station_id;'
pd.read_sql(query, engine.connect())
```
